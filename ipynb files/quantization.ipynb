```python
import os
import tensorflow as tf
from tensorflow.keras.models import load_model
import pathlib

model_dir = r'C:\Users\prema\Downloads\project - Copy\models'

model_files = [f for f in os.listdir(model_dir) if f.endswith(('.keras'))]
models = {f: load_model(os.path.join(model_dir, f)) for f in model_files}

tflite_models_dir = pathlib.Path(r"C:\Users\prema\Downloads\project - Copy\quantization_models")
tflite_models_dir.mkdir(exist_ok=True, parents=True)

for name, model in models.items():
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    converter.target_spec.supported_types = [tf.float16]
    tflite_fp16_model = converter.convert()
    tflite_model_fp16_file = tflite_models_dir / f"{name.replace('.keras', '_fp16.tflite')}"
    tflite_model_fp16_file.write_bytes(tflite_fp16_model)
```
